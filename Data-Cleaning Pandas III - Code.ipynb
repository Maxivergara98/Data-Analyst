```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset_banco.csv')
```

```python
#eliminamos los nulos
df.dropna(inplace = True)
df.info()
```

```python
#contamos los niveles en las diferentes columnas

cols_cat = ['job', 'marital', 'education', 'default','housing', 'loan', 'contact','month','poutcome', 'y']

for col in cols_cat:
    print(f'Columna {col} tiene {df[col].nunique()} niveles' )
```

```python
#eliminamos duplicados
df.drop_duplicates(inplace= True)
print(df.shape)
```

```python
#observar mediante graficos si hay valores numericos anormales.

cols_num = ['age','balance','day','duration','previous']

for col in cols_num:
    plt.figure(figsize = (10,4))
    sns.boxplot(x=df[col])
    plt.title(col)
    plt.show()
```

```python
#eliminamos sujetos con mayores a 100 anios
df = df[df['age'] < 100]
```

```python
#elinamos filas con duration de llamadas negativas
df = df[df['duration'] >0]
```

```python
#eliminamos los previous extremadamentoa altos
df = df[df['previous']<100]
```

```python
#graficar errores tipograficos en columnas categoricas.

cols_cat = ['job', 'marital', 'education', 'default','housing', 'loan', 'contact','month','poutcome', 'y']

for col in cols_cat:
    plt.figure(figsize=(12,4))
    sns.countplot(x=df[col])
    plt.title(col)
    plt.xticks(rotation = 90)
    plt.show()

```

```python
#Pasamos todas las categorias a minuscula

for col in df.columns:
    if col in cols_cat:
        df[col] = df[col].str.lower()
        
for col in cols_cat:
    plt.figure(figsize=(10,4))
    sns.countplot(x=df[col])
    plt.title(col)
    plt.xticks(rotation=90)
    plt.show()
```

```python
#job - Unificamos admin. y Administrative
print(df['job'].unique())
df['job'] = df['job'].str.replace('admin.', 'administrative', regex= False)
print(df['job'].unique())
```

```python
#marital 
```

```python
#marital - Unificamos div. divorced
print(df['marital'].unique())
df['marital'] = df['marital'].str.replace('div.', 'divorced', regex= False)
print(df['marital'].unique())
```

```python
##education - unificamos sec. y secondary, y tambien unk con unknown
print(df['education'].unique())
df['education'] = df['education'].str.replace('sec.','secondary', regex = False)
df[df['education']== 'unk'] ='unknown'
print(df['education'].unique())



```

```python
#contact - unificamos telephone y phone
df['contact'].unique()
df[df['contact'] =='phone'] = 'telephone'
df['contact'].unique()
```

```python
#poutcome: unificar unk con unknown
df['poutcome'].unique()
df[df['poutcome']== 'unk'] = 'unknown'
df['poutcome'].unique()
```
