```python
#extraemos informacion del contracto, si el contracto dice Free o se refiere a On Loan, creamos 3 columnas NA. porque no esta la informacion
def extract_contract_info(contract):
    if contract == 'Free' or 'On Loan' in contract:
        start_date = np.nan
        end_date = np.nan
        contract_lenght = 0
#Dividimos la columna contract en 2, primero registro para start_Date y segundo registro para end_date
#luego start_year le damos los primoers 4 digitos y para end_year, los 4 digitos de end_Date
    else:
        start_date,end_date= contract.split(' ~ ')
        start_year = int(start_date[:4])
        end_year = int(end_date[:4])
        contract_lenght = end_year - start_year
    return start_date, end_date,contract_lenght


#creamos las columnas donde los nuevos valores van a estar
new_cols = ['Contract Start', 'Contract End', 'Contract Lengh- Years']

#aplicamos la funcion sobre los datos de contratos del fifa
new_data = fifa['Contract'].apply(lambda  x: pd.Series(extract_contract_info(x)))


#hacemos un for loopp para poner insertar la informacion junto con sus columnas y valores en el dataframe en una 
#cierta posicion 
for i in range(len(new_cols)):
    fifa.insert(loc=fifa.columns.get_loc('Contract')+1+i,column=new_cols[i], value= new_data[i])




```

```python
#contract Categories:

def status_contract(contract):
    if contract == 'Free':
        return 'Free'
    elif 'On Loan' in contract:
        return 'On Loan'
    else:
        return 'Contract'
    
#add contract status column

fifa.insert(fifa.columns.get_loc('Contract Lengh- Years')+1, 'Contract Status', fifa['Contract'].apply(status_contract))
```

# Player's height in centimeters

```python
fifa['Height'].unique()
```

```python
#convertiro valores en centimetros y eliminar las letras

def convertir_cm (height):
    if 'cm' in height:
        return int(height.strip('cm'))
    else:
        feet, inches = height.split("'")
        total_inches = int(feet)*12 + int(inches.strip('"'))
        return round(total_inches * 2.54)
    
fifa['Height'] = fifa['Height'].apply(convertir_cm)
fifa['Height'].unique()
```

```python
#renombrar columna

fifa = fifa.rename(columns = {'Height': 'Height - cm'})
fifa.sample(1)
```

WEIGHT TO KILOGRAMS


```python
fifa['Weight'].unique()
```

```python
def libras_to_kilos(weight):
    if 'kg' in weight:
        return int(weight.strip('kg'))
    else:
        pounds = int(weight.strip('lbs'))
        return round(pounds / 2.205)
    
#lo aplicamos a los registros de la columna weight mediante la funcion
df['Weight'] = df['Weight'].apply(libras_to_kilos)
df['Weight'].unique()
```

```python
fifa =  fifa.rename(columns = {'Weight': 'Weight-Kilos'})
```
